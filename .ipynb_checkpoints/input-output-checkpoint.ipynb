# Types of joins
1. inner : Take intersection of keys from both dataframes
2. outer : Union of keys
3. left : Only left keys are taken
4. right : Only right keys are taken

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],'value': np.random.randn(4),'key1': ['A', 'B', 'C', 'A']})
df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],'val': np.random.randn(4),'ke': ['A', 'B', 'C', 'D']})
```

```python
print display(df1)
print display(df2)
```

```python
indexed_df2 = df2.set_index('key')
indexed_df2
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1, indexed_df2, left_on='key', right_on='ke')
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K5', 'K6'],'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K1', 'K3'],'C': ['C0', 'C1', 'C2', 'C3'],'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
print display(left)  
print display(right)
```

```python
pd.merge(left,right,how='left',on='key',indicator='True')
```

### Concatinating using append
A useful shortcut to concat are the append instance methods on Series and DataFrame. These methods actually predated concat. They concatenate along axis=0, namely the index:
axis 0 : along rows
axis 1 : along column

NOTE : append return a new dataframe and doesn't modifyoriginal

add row(df) using append

```python
In [1]: df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
   ...:                     'B': ['B0', 'B1', 'B2', 'B3'],
   ...:                     'C': ['C0', 'C1', 'C2', 'C3'],
   ...:                     'D': ['D0', 'D1', 'D2', 'D3']},
   ...:                     index=[0, 1, 2, 3])
   ...: 

In [2]: df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
   ...:                     'B': ['B4', 'B5', 'B6', 'B7'],
   ...:                     'C': ['C4', 'C5', 'C6', 'C7'],
   ...:                     'D': ['D4', 'D5', 'D6', 'D7'],
                            'E': ['D4', 'D5', 'D6', 'D7']},
   ...:                      index=[4, 5, 6, 7])
   ...: 

In [3]: df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
   ...:                     'B': ['B8', 'B9', 'B10', 'B11'],
   ...:                     'C': ['C8', 'C9', 'C10', 'C11'],
   ...:                     'D': ['D8', 'D9', 'D10', 'D11']},
   ...:                     index=[8, 9, 10, 11])
    
type(df2.iloc[0])
```
